**i)Menu Dataset Analysis**

**a)Calories Distribution**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file
menu_df = pd.read_csv('menu.csv', encoding='unicode_escape')
menu_df['Calories'] = pd.to_numeric(menu_df['Calories'], errors='coerce')
menu_df = menu_df.dropna(subset=['Calories'])
counts,bins = np.histogram(menu_df['Calories'],bins=20)
colors = plt.cm.viridis(np.linspace(0,1,len(counts)))
plt.figure(figsize=(8,6))
for count,x,color in zip(counts,bins[:-1],colors):
  plt.bar(x,count,width=np.diff(bins)[0],color=color,edgecolor='black')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.title('Calories Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(menu_df['Total Fat'],menu_df['Calories'],color='blue')
plt.xlabel('Total Fat')
plt.ylabel('Calories')
plt.title('Total Fat vs Calories Scatter Plot')
plt.show()
```

**2. Interactive Treemap for Sales by Category and Sub-Category**

```python
import matplotlib.pyplot as plt
import squarify
import pandas as pd

sale_df = pd.read_csv('Supermart Grocery Sales - Retail Analytics Dataset.csv', encoding='unicode_escape')
category_sales = sale_df.groupby(['Category', 'Sub Category']).agg({'Sales': 'sum'}).reset_index()
labels = category_sales['Category'] + '-' + category_sales['Sub Category']
sizes = category_sales['Sales']
plt.figure(figsize=(10, 8))
squarify.plot(sizes=sizes, label=labels, alpha=.7, color=plt.cm.viridis(sizes/max(sizes)), text_kwargs={'fontsize': 10})
plt.title("Treemap of Sales by Category and Sub-Category")
plt.axis('off')
plt.show()
```

```python
top_category_sales = category_sales.sort_values(by ='Sales', ascending = False).head(10)
labels = top_category_sales['Category'] + '-' + top_category_sales ['Sub Category']
sizes = top_category_sales ['Sales']
plt.figure(figsize=(10, 8))
squarify.plot(sizes=sizes, label=labels, alpha=0.7, color = plt.cm.viridis(sizes / max(sizes)), text_kwargs={'fontsize': 12})
plt.title("Treemap of Top 10 Sales by Category and Sub-Category")
plt.axis('off')
plt.show()
```

**4. Heatmap of Sales and Profit by Region and State**

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
import pandas as pd
sale_df = pd.read_csv('Supermart Grocery Sales - Retail Analytics Dataset.csv', encoding='unicode_escape')
sale_df['Order Date'] = pd.to_datetime(sale_df['Order Date'], format='mixed', errors='coerce')
time_series_data = sale_df.groupby('Order Date')[['Sales', 'Profit']].sum().reset_index()
output_notebook()
source = ColumnDataSource(time_series_data)
p = figure(title="Sales and Profit Over Time", x_axis_label='Order Date', y_axis_label="Amount", x_axis_type="datetime", width=600, height=400)
p.line('Order Date', 'Sales', source=source, legend_label='Sales', line_width=2, color='blue')
p.line('Order Date', 'Profit', source=source, legend_label='Profit', line_width=2, color='green')
p.legend.location = "top_left"
p.legend.click_policy = "hide"
show(p)
```

```python
import altair as alt
import pandas as pd
sale_df = pd.read_csv('Supermart Grocery Sales - Retail Analytics Dataset.csv', encoding='unicode_escape')
sale_df['Order Date'] = pd.to_datetime(sale_df['Order Date'], format='mixed', errors='coerce')
time_series_data = sale_df.groupby('Order Date')[['Sales', 'Profit']].sum().reset_index()
melted_data = time_series_data.melt('Order Date', var_name='Metric', value_name='Amount')
line_chart = alt.Chart(melted_data).mark_line().encode(
    x=alt.X('Order Date:T', title='Order Date'),
    y=alt.Y('Amount:Q', title='Amount'),
    color='Metric:N',
    tooltip=['Metric', 'Amount']
).properties(
    title='Sales and Profit Over Time',
    width=800,
    height=400
).interactive()
line_chart
```

 **2. Interactive Bar Chart for Sales by State**

```python
import altair as alt
import pandas as pd
sale_df = pd.read_csv('Supermart Grocery Sales - Retail Analytics Dataset.csv', encoding='unicode_escape')
sales_by_state = sale_df.groupby('State')['Sales'].sum().reset_index().nlargest(10, 'Sales')
bar_chart = alt.Chart(sales_by_state).mark_bar().encode(
    x=alt.X('Sales:Q', title='Sales'),
    y=alt.Y('State:N', title='State', sort='-x'),
    color='State:N',
    tooltip=['State', 'Sales']
).properties(
    title='Top 10 Sales by State',
    width=600,
    height=400
).interactive()
bar_chart
```
